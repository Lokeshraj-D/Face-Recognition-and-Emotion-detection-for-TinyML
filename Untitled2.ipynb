```python
import os
import numpy as np
import tensorflow as tf
from sklearn.metrics import confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import seaborn as sns

# Define path to test data
test_dir = 'images/images/validation'  # Replace with the actual path to your test directory

# Set image size
image_size = (224, 224)

# Load the model (assumes your model is already trained)
model = tf.keras.models.load_model('fer.keras')  # Replace with the actual model path

# Define a data generator for loading and preprocessing images
test_datagen = ImageDataGenerator(rescale=1./255)  # Normalize images

# Create a test generator to load images from the directory
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=image_size,
    batch_size=32,
    class_mode='categorical',  # For multi-class classification
    shuffle=False  # We need to keep the order of images to compare predictions
)

# Make predictions on the test set
predictions = model.predict(test_generator, verbose=1)
predicted_classes = np.argmax(predictions, axis=1)  # Get the predicted class labels

# Get the true labels from the generator
true_classes = test_generator.classes  # These are the true labels

# Build confusion matrix
cm = confusion_matrix(true_classes, predicted_classes)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=test_generator.class_indices, yticklabels=test_generator.class_indices)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
accuracy = np.sum(np.diag(cm)) / np.sum(cm)  
print(f"Accuracy: {accuracy * 100:.2f}%")
```
