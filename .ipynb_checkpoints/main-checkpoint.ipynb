```python
import tensorflow as tf
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
IMG_SIZE = 224

def produce_dataset():
    dataset = []
    for i in range(1, 5):
        data_path = os.path.join("data", f"n{str(i)}")
        for images in os.listdir(data_path):
            img = cv2.imread(os.path.join(data_path, images))
            dataset.append([img, i])
            '''faces = face_cascade.detectMultiScale(img, 1.3, 5)
            for (x, y, w, h) in faces:
                roi_color = img[y:y+h, x:x+w]
                resized_array = cv2.resize(roi_color, (IMG_SIZE, IMG_SIZE))
                dataset.append([resized_array, i])'''

    print("Length of dataset: ", len(dataset))
    return dataset
```

```python
data = produce_dataset() 
```

```python
X = []
y = []
for feature, label in data:
    X.append(feature)
    y.append(label)
```

```python
samp_ind = np.random.randint(0,601)
plt.imshow(X[samp_ind])
y[samp_ind]
```

```python
X = np.array(X)
X.shape
```

```python
from sklearn.preprocessing import LabelBinarizer
label_bin = LabelBinarizer()
y = label_bin.fit_transform(y)
print (y.shape)
```

```python
y = np.array(y)
print(len(y))
```

```python
from sklearn.model_selection import train_test_split
test_size = 0.10
train_images, test_images, train_labels, test_labels = train_test_split(X, y, random_state=42, test_size=test_size)
print ("train images",len(train_images))
print ("test_images ",len(test_images))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
            rescale=1/255, 
            zoom_range=0.2, 
            width_shift_range=0.3,
            height_shift_range=0.3,
            horizontal_flip=True, 
            rotation_range=30,
            fill_mode = 'nearest')

batch_size = 16

train_generator = datagen.flow(train_images, 
                               train_labels, 
                               batch_size=batch_size)

print ("train images generator",len(train_generator))
```

```python
from sklearn import preprocessing

test_images = np.array(test_images)

min_max_scaler = preprocessing.MinMaxScaler()
test_images_minmax = min_max_scaler.fit_transform(test_images.reshape(-1, test_images.shape[-1])).reshape(test_images.shape)
print (len(test_images_minmax))
```

```python
from tensorflow.keras.layers import Input, Lambda, Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.layers import BatchNormalization, Dropout, GlobalAveragePooling2D, Activation
from tensorflow.keras.models import Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications import MobileNetV3Small
```

```python
mobilenet_model = MobileNetV3Small(
    input_shape = (224, 224, 3),
    alpha=1.0,
    #minimalistic=False,
    include_top=False,
    weights="imagenet",
    #dropout_rate=0.2,
    #classifier_activation="softmax",
    include_preprocessing=False,
)
```

```python
def custom_layers(bottom_model, num_class):
    top_model = bottom_model.output
    top_model = GlobalAveragePooling2D()(top_model)
    #top_model = Dense(1024,activation='relu')(top_model)
    #top_model = Dropout(0.15)(top_model)
    top_model = Dense(1024,activation='relu')(top_model)
    top_model = Dropout(0.05)(top_model)
    top_model = Dense(512,activation='relu')(top_model)
    top_model = BatchNormalization()(top_model)
    top_model = Dense(256,activation='relu')(top_model)
    top_model = Dense(num_class,activation='softmax')(top_model)
    return top_model
```

```python
num_class = 4
top_model = custom_layers(mobilenet_model, num_class)
model = Model(inputs = mobilenet_model.input, outputs = top_model)
model.summary()
```

```python
optimizer1 = tf.keras.optimizers.Adam(
    learning_rate=0.001,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07,
    amsgrad=True,
    name='Adam',
)
model.compile(loss="categorical_crossentropy",
              optimizer= optimizer1,
              metrics=["accuracy"])
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping

checkpoint = ModelCheckpoint("face_recogV2.keras",
                             monitor="val_loss",
                             mode="min",
                             save_best_only = True,
                             verbose=1)

earlystop = EarlyStopping(monitor = 'val_loss', 
                          min_delta = 0, 
                          patience = 10,
                          verbose = 1,
                          restore_best_weights = True)

callbacks = [earlystop, checkpoint]
```

```python
import time

start = time.time()
epochs = 50
#train_generator = train_generator.repeat()

history = model.fit(train_generator, 
                    steps_per_epoch = len(train_images) // batch_size,
                    epochs=epochs,  
                    callbacks = callbacks,
                    shuffle = True,
                    validation_data= (test_images_minmax, test_labels), 
                    validation_steps = len(test_images_minmax))
print("Total time: ", time.time() - start, "seconds")
```

```python
test_loss, test_accuracy = model.evaluate(test_images_minmax, test_labels, verbose=1)
print('Test accuracy :', test_accuracy)
print('Test loss :', test_loss)

print('Model accuracy is %.2f%% and Loss accuracy is %.2f%%' % (test_accuracy, test_loss))
```

```python
cam = cv2.VideoCapture(0)
import matplotlib.pyplot as plt

target_fps = 10
while cam.isOpened():
    ret, frame = cam.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
    for (x,y,w,h) in faces:
        #print("check")
        cv2.rectangle(frame, (x,y), (x+w, y+h), (255, 0, 0), 2)
        input_im = gray[y:y+h, x:x+w]
        input_im = cv2.merge([input_im, input_im, input_im])
        input_original = input_im.copy()
        input_original = cv2.resize(input_original, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_LINEAR)
        input_im = cv2.resize(input_im, (224, 224), interpolation = cv2.INTER_LINEAR)
        input_im = input_im / 255.
        input_im = input_im.reshape(1,224,224,3)
        res = np.argmax(model.predict(input_im, 1, verbose = 0), axis=1)
        #txt = f'{w} x {h}'
        cv2.putText(frame, 
                data_dict[str(res+1)],
                (x, y-10),
                fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                fontScale=0.6,
                color=(255, 0, 0),  # BGR color for the text
                thickness=2
            )
    cv2.imshow('image', frame)
    if cv2.waitKey(1) and 0XFF == ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```

```python
import os

train_data_dir = 'data/'

def process_images_in_folder(folder_path):
    for person_folder in os.listdir(folder_path):
        person_path = os.path.join(folder_path, person_folder)

        print(f"Processing folder: {person_folder}")

        # Loop through each image in the person's folder
        for image_name in os.listdir(person_path):
            image_path = os.path.join(person_path, image_name)

            # Check if it is an image file
            if image_path.lower().endswith(('.jpg')):
                image = cv2.imread(image_path)
                if image is None:
                    continue
                gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
                if len(faces)==0:
                    continue

                for (x, y, w, h) in faces:      
                    img = image[y:y+h, x:x+w]
                    img = cv2.resize(img, (224, 224))

                    cv2.imwrite(image_path, img)
            cv2.waitKey(100)
        
            
process_images_in_folder(train_data_dir)

# Release all windows
cv2.destroyAllWindows()
```
