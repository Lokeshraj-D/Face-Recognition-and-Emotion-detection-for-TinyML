```python
from keras.models import load_model

face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
model = load_model('RP.keras')
```

```python

cam = cv2.VideoCapture(0)

while True:
    #cam.set(cv2.CAP_PROP_FPS,100) 
    ret, frame = cam.read()
    #if cv2.waitKey(1) & 0XFF == ord('c'):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        img = frame[y:y+h, x:x+w]
        frame = cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        img = np.array(img)
        img = cv2.resize(img, (224, 224), interpolation=cv2.INTER_NEAREST)
        #mms = MinMaxScaler()
        #img = mms.fit_transform(img.reshape(-1, img.shape[-1])).reshape(img.shape)
        img = img/255.
        img = img.reshape(1,224,224,3)
        res = np.argmax(model.predict(img, 1, verbose = 0), axis=1)
        
        pred = dict_emo[res[0]]
        cv2.putText(frame, 
                pred,
                (x, y-10),
                fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                fontScale=0.6,
                color=(255, 0, 0),  # BGR color for the text
                thickness=2
            )
            #sleep(3)
            
    cv2.imshow("Image: " ,frame)
    if cv2.waitKey(1) & 0XFF == ord('t'):
        break

cam.release()
cv2.destroyAllWindows()
    
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import model_from_json
import copy


model_json_file = "model.json"
model_weights_file = "emotion_detect.h5"
with open(model_json_file, "r") as json_file:
    loaded_model_json = json_file.read()
    classifier = model_from_json(loaded_model_json)
    classifier.load_weights(model_weights_file)


cap = cv2.VideoCapture(0)


while True:
    ret, frame = cap.read()
    img1 = copy.deepcopy(frame)
    gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        fc = gray[y:y+h, x:x+w]
        img = frame[y:y+h, x:x+w]
        img = np.array(img)
        img = cv2.resize(img, (224, 224), interpolation=cv2.INTER_NEAREST)
        roi = cv2.resize(fc, (48,48))
        pred = classifier.predict(roi[np.newaxis, :, :, np.newaxis])
        img = img/255.
        img = img.reshape(1,224,224,3)
        res = np.argmax(model.predict(img, 1, verbose = 0), axis=1)
        text_idx=np.argmax(pred)
        pred_name = dict[res[0]]
        text_list = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']
        if text_idx == 0:
            text= text_list[0]
        if text_idx == 1:
            text= text_list[1]
        elif text_idx == 2:
            text= text_list[2]
        elif text_idx == 3:
            text= text_list[3]
        elif text_idx == 4:
            text= text_list[4]
        elif text_idx == 5:
            text= text_list[5]
        elif text_idx == 6:
            text= text_list[6]

        op = pred_name + "   " + text
        cv2.putText(img1, op, (x, y-5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.45, (255, 0, 255), 2)
        img1 = cv2.rectangle(img1, (x,y), (x+w, y+h), (0,0,255), 2)


    cv2.imshow("Face Detection:", img1)
    key = cv2.waitKey(1) & 0xFF
    if key== ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
